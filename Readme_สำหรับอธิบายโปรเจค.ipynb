#ไฟล์ที่ 1 Re ไฟล์ที่ื 1 สำหรับการ Recolumn Etl Pipeline Duckdb แรกเพื่อทำ Dimentinal ของ ข้อมูล

```python
#ไฟล์ที่ื 1 สำหรับการ Recolumn Etl Pipeline Duckdb แรกเพื่อทำ Dimentinal ของ ข้อมูล

import duckdb as dd
import polars as pl
import pandas as pd
import dlt
import sqlite3 as sq
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import streamlit as st

def recolname(df, namedict):
    return df.rename(namedict)

conn = sq.connect(r'C:\Users\ASUS\mini\chinook.db')

#Recolumn name for make stg
albums = (
    pl.read_database(
   query= "SELECT * FROM albums",
   connection=conn )
   .pipe(recolname, {'AlbumId': 'Album_ID'})
   .pipe(recolname, {'ArtistId': 'Artist_ID '})
   .pipe(recolname, {'Title': 'AL_title'})
)

artists =(
    pl.read_database(
   query= "SELECT * FROM artists",
   connection=conn )
   .pipe(recolname, {'ArtistId': 'Artist_ID'})
   .pipe(recolname, {'Name': 'AR_Name'})
)
pl.Config.set_tbl_cols(-1)
customers = (
    pl.read_database(
        query="SELECT * FROM customers",
        connection=conn)
    .pipe(recolname, {'CustomerId': 'Customer_ID'})
    .pipe(recolname, {'PostalCode': 'Cus_PostalCode'})
    .pipe(recolname, {'State': 'Cus_State'})
    .pipe(recolname, {'FirstName': 'Cus_FirstName'})
    .pipe(recolname, {'City': 'Cus_City'})
    .pipe(recolname, {'LastName': 'Cus_LastName'})
    .pipe(recolname, {'Fax': 'Cus_Fax'})
    .pipe(recolname, {'Email': 'Cus_Email'})
    .pipe(recolname, {'Phone': 'Cus_Phone'})
    .pipe(recolname, {'Country': 'Cus_Country'})
    .pipe(recolname, {'Address': 'Cus_Address'})
    .pipe(recolname, {'SupportRepId': 'Employee_ID'})
)


employees = (
   pl.read_database(
        query="SELECT * FROM employees",
        connection=conn)
    .pipe(recolname, {'EmployeeId': 'Employee_ID'})
    .pipe(recolname, {'Title': 'EM_title'})
    .pipe(recolname, {'PostalCode': 'EM_PostalCode'})
    .pipe(recolname, {'State': 'EM_State'})
    .pipe(recolname, {'FirstName': 'EM_FirstName'})
    .pipe(recolname, {'City': 'EM_City'})
    .pipe(recolname, {'LastName': 'EM_LastName'})
    .pipe(recolname, {'Fax': 'EM_Fax'})
    .pipe(recolname, {'Email': 'EM_Email'})
    .pipe(recolname, {'Phone': 'EM_Phone'})
    .pipe(recolname, {'Country': 'EM_Country'})
    .pipe(recolname, {'Address': 'EM_Address'})
    .pipe(recolname, {'ReportsTo': 'EM_ReportsTo'})
    .pipe(recolname, {'BirthDate': 'EM_BirthDate'})
    .pipe(recolname, {'HireDate': 'EM_HireDate'})
)

#set Datetime & Hiredate To Year and Month
employees = employees.with_columns(
    pl.col('EM_BirthDate').str.strptime(pl.Datetime, format="%Y-%m-%d %H:%M:%S", strict=False)
)

employees = employees.with_columns(
    pl.col('EM_HireDate').str.strptime(pl.Datetime, format="%Y-%m-%d %H:%M:%S", strict=False)
)

#Set Code Col for Do column
col = employees["EM_BirthDate"]
col1 = employees["EM_HireDate"]


genres =(pl.read_database(
   query= "SELECT * FROM genres",
   connection=conn )
   .pipe(recolname, {'GenreId': 'Genre_ID'})
   .pipe(recolname, {'Name': 'Genre_Name'})
)

invoices =(
    pl.read_database(
        query="SELECT * FROM invoices",
        connection=conn)
    .pipe(recolname, {'BillingPostalCode': 'IN_PostalCode'})
    .pipe(recolname, {'InvoiceId': 'Invoice_ID'})
    .pipe(recolname, {'CustomerId': 'Customer_ID'})
    .pipe(recolname, {'InvoiceDate': 'IN_InvoiceDate'})
    .pipe(recolname, {'BillingAddress': 'IN_BillingAddress'})
    .pipe(recolname, {'BillingCity': 'IN_BillingCity'})
    .pipe(recolname, {'BillingState': 'IN_BillingState'})
    .pipe(recolname, {'BillingCountry': 'IN_BillingCountry'})
    .pipe(recolname, {'Total': 'IN_Total'})
)
invoices = invoices.with_columns(
    pl.col('IN_InvoiceDate').str.strptime(pl.Datetime, format="%Y-%m-%d %H:%M:%S", strict=False)
)





invoice_items =(
    pl.read_database(
   query= "SELECT * FROM invoice_items",
   connection=conn )
   .pipe(recolname, {'InvoiceLineId': 'Invoiceline_ID'})
   .pipe(recolname, {'InvoiceId': 'Invoice_ID'})
   .pipe(recolname, {'TrackId': 'Track_ID'})
   .pipe(recolname, {'UnitPrice': 'II_UnitPrice'})
   .pipe(recolname, {'Quantity': 'II_Quantity'})
)

media_types =(
    pl.read_database(
   query= "SELECT * FROM media_types",
   connection=conn )
   .pipe(recolname, {'MediaTypeId': 'MediaType_ID'})
   .pipe(recolname, {'Name': 'MediaType_Name'})
)

playlists =(
    pl.read_database(
   query= "SELECT * FROM playlists",
   connection=conn )
   .pipe(recolname, {'PlaylistId': 'Playlist_ID'})
   .pipe(recolname, {'Name': 'Pl_Name'})
)

playlist_track =(
   pl.read_database(
   query= "SELECT * FROM playlist_track",
   connection=conn )
   .pipe(recolname, {'PlaylistId': 'Playlist_ID'})
   .pipe(recolname, {'TrackId': 'Track_ID'})
)

tracks =(
    pl.read_database(
   query= "SELECT * FROM tracks",
   connection=conn )
   .pipe(recolname, {'TrackId': 'Track_ID'})
   .pipe(recolname, {'MediaTypeId': 'MediaType_ID'})
   .pipe(recolname, {'AlbumId': 'Album_ID'})
   .pipe(recolname, {'GenreId': 'Genre_ID'})
   .pipe(recolname, {'Name': 'TR_Name'})
   .pipe(recolname, {'Composer': 'TR_Composer'})
   .pipe(recolname, {'Milliseconds': 'TR_Milliseconds'})
   .pipe(recolname, {'Bytes': 'TR_Bytes'})
   .pipe(recolname, {'UnitPrice': 'TR_UnitPrice'})
)

albums = albums.to_pandas()
artists = artists.to_pandas()
customers = customers.to_pandas()
employees = employees.to_pandas()
genres = genres.to_pandas()
invoices = invoices.to_pandas()
invoice_items = invoice_items.to_pandas()
media_types = media_types.to_pandas()
playlists = playlists.to_pandas()
playlist_track = playlist_track.to_pandas()
tracks = tracks.to_pandas()

pipeline =dlt.pipeline(
    pipeline_name="mini_Retry_project",destination="duckdb",
     dataset_name="Chin"
)

#set สำหรับ pipe line ไปเพื่อให้ ไฟล์ mini_Retry_project เป็น duckdb สำหรับทำ Stagging table และชื่อ Chin
pipeline.run(albums, table_name="stg_albums",write_disposition="replace")
pipeline.run(artists, table_name="stg_artists",write_disposition="replace")
pipeline.run(customers, table_name="stg_customers",write_disposition="replace")
pipeline.run(employees, table_name="stg_employees",write_disposition="replace")
pipeline.run(genres, table_name="stg_genres",write_disposition="replace")
pipeline.run(invoice_items, table_name="stg_invoice_items",write_disposition="replace")
pipeline.run(invoices, table_name="stg_invoices",write_disposition="replace")
pipeline.run(media_types, table_name="stg_media_types",write_disposition="replace")
pipeline.run(playlists, table_name="stg_playlists",write_disposition="replace")
pipeline.run(playlist_track, table_name="stg_playlist_track",write_disposition="replace")
pipeline.run(tracks, table_name="stg_tracks",write_disposition="replace")
```

#ไฟล์ที่ 2 ดึงข้อมูลจาก Duckdb ใหม่ที่สร้างขึ้นและสร้าง Dimenstional และ ทดสอบ Datacube

```python
#ไฟล์ที่ 2 ดึงข้อมูลจาก Duckdb ใหม่ที่สร้างขึ้นและสร้าง Dimenstional และ ทดสอบ Datacube

import duckdb as dd
import polars as pl
import pandas as pd
from datetime import datetime
import dlt

def sort_by(df, bycolname):
    return df.with_columns(pl.col(bycolname).sort())

def get_unique(df, bycolname):
    return df.unique(bycolname)

def remove_null(df, colname):
    return df.filter(pl.col(colname).is_not_null())

def recolname(df, namedict):
    return df.rename(namedict)

#Connect to the DuckDB database
name_db ='mini_retry_project.duckdb'
conn =dd.connect(name_db)
pl.Config.set_tbl_cols(-1)



dim_albums    = pl.read_database(query="select * from Chin.stg_albums",connection=conn)
dim_artists   = pl.read_database(query="select * from Chin.stg_artists",connection=conn)
dim_customers = pl.read_database(query="select * from Chin.stg_customers",connection=conn)
dim_employees = pl.read_database(query="select * from Chin.stg_employees",connection=conn)
dim_genres    = pl.read_database(query="select * from Chin.stg_genres",connection=conn)
dim_invoice_items = pl.read_database(query="select * from Chin.stg_invoice_items",connection=conn)
dim_invoices      = pl.read_database(query="select * from Chin.stg_invoices",connection=conn)
dim_media_types   = pl.read_database(query="select * from Chin.stg_media_types",connection=conn)
dim_playlists = pl.read_database(query="select * from Chin.stg_playlists",connection=conn)
dim_playlist_track= pl.read_database(query="select * from Chin.stg_playlist_track",connection=conn)
dim_tracks        = pl.read_database(query="select * from Chin.stg_tracks",connection=conn)

conn.close()

dim_invoices = dim_invoices.with_columns([
    pl.col('in_invoice_date').dt.year().alias('year'),
    pl.col('in_invoice_date').dt.month().alias('month'),
    pl.col('in_invoice_date').dt.day().alias('day'),
    pl.col('in_invoice_date').dt.quarter().alias('quarter')
])

dim_invoices_invoice_items =dim_invoices.join(dim_invoice_items, on='invoice_id', how='left')

print(dim_invoices.dtypes)

print(dim_invoices.columns)

print(dim_invoices_invoice_items.columns)
print(dim_albums.columns)
print(dim_tracks.columns)

fact_sale = (dim_invoices_invoice_items
               .join(dim_tracks, on='track_id', how='left')
               .join(dim_albums, on='album_id', how='left')
               .join(dim_media_types, on='media_type_id', how='left')
               .join(dim_genres, on='genre_id', how='left')
               .join(dim_customers, on='customer_id', how='left')
               .join(dim_customers, on='employee_id', how='left')
               )

fact_sale  = fact_sale [['track_id','album_id','media_type_id','genre_id',
                                 'customer_id','employee_id','ii_unit_price','ii_quantity','day','month','year','quarter']]

dim_tracks = dim_tracks[['track_id','tr_name']]

dim_tracks

dim_customers = dim_customers[['customer_id','cus_first_name','cus_last_name','company',
                               'cus_fax','cus_email','cus_country','cus_city','employee_id']]

dim_customers.columns

dim_employees  = dim_employees [['employee_id','em_first_name','em_last_name','em_title','em_birth_date',
                                 'em_reports_to','em_phone','em_fax','em_email']]
#join albumและ artist เข้าด้วยกันเพื่อนำไปทำ datacube

dim_artists_albums = dim_artists.join(dim_albums, on='artist_id', how='left')

dim_artists_albums = dim_artists_albums.with_columns([
    pl.col('album_id').cast(pl.Int64)
])


#joinเพื่อต้องการให้ทราบเพิ่มว่า tackอยู่ใน platlist ชื่ออะไร
dim_playlist_track_playlists = dim_playlists.join(dim_playlist_track, on='playlist_id', how='left')
dim_playlist_track_playlists_tracks = dim_tracks.join(dim_playlist_track_playlists, on='track_id', how='left')


dim_customers = dim_customers.with_columns(
    (pl.col('cus_first_name') + ' ' + pl.col('cus_last_name')).alias('cus_name')
)
dim_employees = dim_employees.with_columns(
    (pl.col('em_first_name') + ' ' + pl.col('em_last_name')).alias('em_name')
)


# สร้าง dim_date จากข้อมูล invoice_date ใน dim_invoices
dim_date = dim_invoices.select([
    pl.col('in_invoice_date').alias('full_date'),
    pl.col('year').alias('year'),                  # ปี
    pl.col('month').alias('month'),                # เดือน
    pl.col('day').alias('day'),                    # วัน
    pl.col('quarter').alias('quarter')             # ไตรมาส
]).unique().with_columns(
    pl.arange(0, pl.len()).alias('date_id')      # เพิ่มคอลัมน์ date_id เป็น primary key สำหรับ dim_date
)

fact_sale = fact_sale.join(dim_date, on=['year', 'month', 'day', 'quarter'], how='left')

print(dim_date)

print(fact_sale.columns)
print(dim_date.columns)


#รันทดสอบ code Datacube ไม่จำเป็นต้องรันก็ได้
datacube_tracks =(fact_sale.select(pl.col('track_id','genre_id','media_type_id','album_id','customer_id','employee_id',
                                            'ii_unit_price', 'ii_quantity','date_id','day','month','year','quarter'))
            .join(dim_playlist_track_playlists_tracks,on='track_id',how='left')
            .join(dim_genres,on='genre_id',how='left')
            .join(dim_media_types,on='media_type_id',how='left')
            .join(dim_artists_albums,on='album_id',how='left')
            .join(dim_customers,on='customer_id',how='left')
            .join(dim_employees,on='employee_id',how='left')
            .join(dim_date,on='date_id',how='left')
            )

datacube_tracks


#แปลงเป็น Pandas เพือ pipeline
dim_playlist_track_playlists_tracks = dim_playlist_track_playlists_tracks.to_pandas()
fact_sale = fact_sale.to_pandas()
dim_genres = dim_genres.to_pandas()
dim_media_types = dim_media_types.to_pandas()
dim_artists_albums = dim_artists_albums.to_pandas()
dim_customers=dim_customers.to_pandas()
dim_employees=dim_employees.to_pandas()
dim_date=dim_date.to_pandas()

pipeline =dlt.pipeline(
    pipeline_name="mini_retry_project2",destination="duckdb",
     dataset_name="Chin2"
)

pipeline.run(dim_playlist_track_playlists_tracks, table_name="dim_pltrack_playlists_tr",write_disposition="replace")
pipeline.run(fact_sale, table_name="fact_sale",write_disposition="replace")
pipeline.run(dim_customers, table_name="dim_cus",write_disposition="replace")
pipeline.run(dim_employees, table_name="dim_emp",write_disposition="replace")
pipeline.run(dim_genres, table_name="dim_genres",write_disposition="replace")
pipeline.run(dim_media_types, table_name="dim_media_types",write_disposition="replace")
pipeline.run(dim_artists_albums, table_name="dim_artists_albums",write_disposition="replace")
pipeline.run(dim_date, table_name="dim_date",write_disposition="replace")
```

#ไฟล์ที่ 3 สำหรับการทำ Dashboard และ Web app

```python
#ไฟล์ที่ 3 สำหรับการทำ Dashboard และ Web app

import duckdb as dd
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
import pandas as pd
import seaborn as sns


pl.Config.set_tbl_cols(-1)

name_db1 = 'mini_retry_project2.duckdb'
conn = dd.connect(name_db1)


dim_pltrack_playlists_tr  = pl.read_database(query="select * from Chin2.dim_pltrack_playlists_tr",connection=conn)
fact_sale   = pl.read_database(query="select * from Chin2.fact_sale",connection=conn)
dim_cus = pl.read_database(query="select * from Chin2.dim_cus",connection=conn)
dim_emp = pl.read_database(query="select * from Chin2.dim_emp",connection=conn)
dim_genres    = pl.read_database(query="select * from Chin2.dim_genres",connection=conn)
dim_media_types = pl.read_database(query="select * from Chin2.dim_media_types",connection=conn)
dim_artists_albums    = pl.read_database(query="select * from Chin2.dim_artists_albums",connection=conn)
dim_date_time    = pl.read_database(query="select * from Chin2.dim_date",connection=conn)


dim_artists_albums = dim_artists_albums.with_columns([
    pl.col('album_id').cast(pl.Int64).alias('album_id')
])


conn.commit()
conn.close()

datacube_tracks =(fact_sale.select(pl.col('track_id','genre_id','media_type_id','album_id','customer_id','employee_id',
                                            'ii_unit_price', 'ii_quantity','date_id','day','month','year','quarter'))
            .join(dim_pltrack_playlists_tr,on='track_id',how='left')
            .join(dim_genres,on='genre_id',how='left')
            .join(dim_media_types,on='media_type_id',how='left')
            .join(dim_artists_albums,on='album_id',how='left')
            .join(dim_cus,on='customer_id',how='left')
            .join(dim_emp,on='employee_id',how='left')
            .join(dim_date_time,on='date_id',how='left')
            )



# Display datacube in Streamlit
st.title("Music Sales Data Exploration")

st.header("Data Overview Dashboard")
st.write("This is a preview of the joined data (datacube).")
st.dataframe(datacube_tracks.head(200000))

# Allow users to filter and visualize the data by Genre
st.title("This is Our question 16 Question")

# 1. ยอดขายรวมตามปีและไตรมาส
st.header("1.Total Sales by Year and Quarter")
rollup_by_year_quarter = (
    datacube_tracks
    .group_by(['year', 'quarter'])
    .agg([
        pl.col('ii_unit_price').sum().alias('total_sales'),
        pl.col('ii_quantity').sum().alias('total_quantity')
    ])
    .sort('total_quantity', descending=True).head(20)
)
st.dataframe(rollup_by_year_quarter.to_pandas())

fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(data=rollup_by_year_quarter, x='quarter', y='total_quantity',hue='year', ax=ax)

plt.title("Total Sales by Year and Quarter")
plt.xlabel("quarter")
plt.ylabel("total_quantity")
st.pyplot(fig)


# 2. ยอดขายรวมตามประเภทเพลง
st.header("2.Total Sales by Genre Top5")
sales_by_genre = (
    datacube_tracks
    .group_by(['genre_id','genre_name'])
    .agg([
        pl.col('ii_unit_price').sum().alias('total_sales')
    ])
    .sort('total_sales', descending=True).head(5)
)
st.dataframe(sales_by_genre.to_pandas())
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(data=sales_by_genre, x='genre_name', y='total_sales',hue='genre_name', ax=ax)

plt.title("Total Sales by Genre Top5")
plt.xlabel("genre_name")
plt.ylabel("total_sales")
st.pyplot(fig)

# 3. ยอดขายรวมแต่ละปี
st.header("3.Total Sales for January to December Across All Years")
filtered_data = datacube_tracks.filter(pl.col('month').is_between(1, 12))
total_sales_by_year = (
    filtered_data
    .group_by(['year'])  # จัดกลุ่มตามปี
    .agg([
        pl.col('ii_unit_price').sum().alias('total_sales')  # รวมยอดขายทั้งหมดต่อปี
    ])
    .sort('year').head(5)  # เรียงลำดับตามปี
)
st.dataframe(total_sales_by_year.to_pandas())
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(data=total_sales_by_year, x='year', y='total_sales',hue='year', ax=ax)

plt.title("Total Sales for January to December Across All Years")
plt.xlabel("year")
plt.ylabel("total_sales")
st.pyplot(fig)

# 4. จำนวนการขายตามประเภทสื่อ
st.header("4.Total Quantity Sold by Media Type")
quantity_by_media_type = (
    datacube_tracks
    .group_by(['media_type_id','media_type_name'])
    .agg([
        pl.col('ii_quantity').sum().alias('total_quantity')
    ])
    .sort('total_quantity', descending=True)
)
st.dataframe(quantity_by_media_type.to_pandas())
fig, ax = plt.subplots(figsize=(8, 12))
sns.barplot(data=quantity_by_media_type, y='media_type_name', x='total_quantity',hue='media_type_name', ax=ax)

plt.title("Total Quantity Sold by Media Type")
plt.ylabel("media_type_name")
plt.xlabel("total_quantity")
st.pyplot(fig)

# 5. จำนวนการขายต่ออัลบัม
st.header("5.Total Quantity Sold by album top 10")
quantity_by_artist = (
    datacube_tracks
    .group_by(['album_id','al_title'])
    .agg([
        pl.col('ii_quantity').sum().alias('total_quantity')
    ])
    .sort('total_quantity', descending=True).head(10)
)
st.dataframe(quantity_by_artist.to_pandas())
fig, ax = plt.subplots(figsize=(8, 8))
sns.barplot(data=quantity_by_artist, y='al_title', x='total_quantity',hue='al_title', ax=ax)

plt.title("Total Quantity Sold by Media Type")
plt.ylabel("al_title")
plt.xlabel("total_quantity")
st.pyplot(fig)

# 6. ยอดขายตามลูกค้า
st.header("6.Total Sales by Customer")
sales_by_customer = (
    datacube_tracks
    .group_by(['customer_id','cus_name','cus_email','cus_country'])
    .agg([
        pl.col('ii_unit_price').sum().alias('total_sales')
    ])
    .sort('total_sales', descending=True).head(10)
)
st.dataframe(sales_by_customer.to_pandas())
fig, ax = plt.subplots(figsize=(8,8))
sns.barplot(data=sales_by_customer, y='cus_name', x='total_sales',hue='cus_name', ax=ax)

plt.title("Total Sales by Customer")
plt.ylabel("cus_name")
plt.xlabel("total_sales")
st.pyplot(fig)

# ึ7.Top 10 Country Most sale out
st.header("7.Top 10 Country Most sale out")
sales_by_country = (
    datacube_tracks
    .group_by(['cus_country'])
    .agg([
        pl.col('ii_unit_price').sum().alias('total_sales')
    ])
    .sort('total_sales', descending=True).head(10)
)
st.dataframe(sales_by_country.to_pandas())
fig, ax = plt.subplots(figsize=(8, 8))
sns.barplot(data=sales_by_country, y='cus_country', x='total_sales',hue='cus_country', ax=ax)

plt.title("Top 10 Country Most sale out")
plt.ylabel("cus_country")
plt.xlabel("total_sales")
st.pyplot(fig)

# 8.Main employee sale out Count
st.header("8.Main employee sale out Count")
Best_employee_sale = (
    datacube_tracks
    .group_by('employee_id_right','em_name')
    .agg([
        pl.col('ii_unit_price').sum().alias('total_sales')
    ])
    .sort('total_sales', descending=True).head()
)
st.dataframe(Best_employee_sale.to_pandas())
fig, ax = plt.subplots(figsize=(8, 8))
sns.barplot(data=Best_employee_sale, x='em_name', y='total_sales',hue='em_name', ax=ax)

plt.title("Total Quantity Sold by Media Type")
plt.xlabel("em_name")
plt.ylabel("total_sales")
st.pyplot(fig)


# 9.Average Monthly Sales by Media Type
st.header("9.Top Monthly quantity Sales by artist")
avg_quantity_sales_by_artist = (
    datacube_tracks
    .group_by(['artist_id', 'ar_name','year'])
    .agg([
        pl.col('ii_quantity').sum().alias('total_quantity'),
        pl.col('ii_unit_price').sum().alias('total_sales')
    ])
    .sort('total_quantity','total_sales', descending=True).head(10)
)
st.dataframe(avg_quantity_sales_by_artist.to_pandas())

# 10.Annual Sales Trend by Artist
st.header("10.Annual quantity Trend by Artist")
sales_trend_by_artist = (
    datacube_tracks
    .group_by(['album_id', 'year','artist_id','ar_name','genre_name','media_type_name'])
    .agg([
        pl.col('ii_quantity').sum().alias('total_quantity'),
        pl.col('ii_unit_price').sum().alias('total_sales')
    ])
    .sort('total_quantity','total_sales', descending=True).head(10)
)
st.dataframe(sales_trend_by_artist.to_pandas())
fig, ax = plt.subplots(figsize=(8, 8))
sns.barplot(data=sales_trend_by_artist, x='total_quantity', y='ar_name',hue='ar_name', ax=ax)

plt.title("Annual quantity Trend by Artist")
plt.xlabel("total_quantity")
plt.ylabel("ar_name")
st.pyplot(fig)

# 11.Most popular track saleing of all time
st.header("11.Most popular track saleing of all time Top 50!")
Most_popular_track = (
    datacube_tracks
    .group_by(['track_id', 'tr_name','artist_id','ar_name','genre_name','album_id'])
    .agg([
        pl.col('ii_unit_price').sum().alias('total_sales')
    ])
    .sort('total_sales', descending=True).head(50)
)
st.dataframe(Most_popular_track.to_pandas())

st.header("12.Best Selling Genre in Q1")
best_genre_q1 = (
    datacube_tracks
    .filter((pl.col('quarter') == 1) )
    .group_by('genre_id','genre_name')
    .agg([
        pl.col('ii_unit_price').sum().alias('total_sales')
    ])
    .sort('total_sales', descending=True)
)
st.dataframe(best_genre_q1.to_pandas())
fig, ax = plt.subplots(figsize=(8, 12))
sns.barplot(data=best_genre_q1, y='genre_name', x='total_sales', hue='genre_id', ax=ax)

plt.title("Total Sales by Genre in Each Quarter")
plt.ylabel("Genre")
plt.xlabel("Total Sales")
plt.legend(title='Genre ID')
st.pyplot(fig)


st.header("13.Best Selling Genre in Q2")
best_genre_q2 = (
    datacube_tracks
    .filter((pl.col('quarter') == 2) )
    .group_by('genre_id','genre_name')
    .agg([
        pl.col('ii_unit_price').sum().alias('total_sales')
    ])
    .sort('total_sales', descending=True)
)
st.dataframe(best_genre_q2.to_pandas())
fig, ax = plt.subplots(figsize=(8, 12))
sns.barplot(data=best_genre_q2, y='genre_name', x='total_sales', hue='genre_id', ax=ax)

plt.title("Total Sales by Genre in Each Quarter")
plt.ylabel("Genre")
plt.xlabel("Total Sales")
plt.legend(title='Genre ID')
st.pyplot(fig)


st.header("14.Best Selling Genre in Q3")
best_genre_q3 = (
    datacube_tracks
    .filter((pl.col('quarter') == 3) )
    .group_by('genre_id','genre_name')
    .agg([
        pl.col('ii_unit_price').sum().alias('total_sales')
    ])
    .sort('total_sales', descending=True)
)
st.dataframe(best_genre_q3.to_pandas())
fig, ax = plt.subplots(figsize=(8, 12))
sns.barplot(data=best_genre_q3, y='genre_name', x='total_sales', hue='genre_id', ax=ax)

plt.title("Total Sales by Genre in Each Quarter")
plt.ylabel("Genre")
plt.xlabel("Total Sales")
plt.legend(title='Genre ID')
st.pyplot(fig)


st.header("15.Best Selling Genre in Q4")
best_genre_q4 = (
    datacube_tracks
    .filter((pl.col('quarter') == 4) )
    .group_by('genre_id','genre_name')
    .agg([
        pl.col('ii_unit_price').sum().alias('total_sales')
    ])
    .sort('total_sales', descending=True)
)
st.dataframe(best_genre_q4.to_pandas())
fig, ax = plt.subplots(figsize=(8, 12))
sns.barplot(data=best_genre_q4, y='genre_name', x='total_sales', hue='genre_id', ax=ax)

plt.title("Total Sales by Genre in Each Quarter")
plt.ylabel("Genre")
plt.xlabel("Total Sales")
plt.legend(title='Genre ID')
st.pyplot(fig)

st.header("16.sales_by_genre_all_quarters graph")
sales_by_genre_all_quarters = (
    datacube_tracks
    .group_by(['genre_id', 'quarter','genre_name'])
    .agg([
        pl.col('ii_unit_price').sum().alias('total_sales')
    ])
    .sort(['genre_id', 'quarter'])
)

# แปลงข้อมูลจาก Polars เป็น Pandas
sales_df = sales_by_genre_all_quarters.to_pandas()
st.dataframe(sales_df)
# พล็อตกราฟแสดงยอดขายตามไตรมาสสำหรับแต่ละประเภทเพลง
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(data=sales_df, x='quarter', y='total_sales', hue='genre_id', ax=ax)

plt.title("Total Sales by Genre in Each Quarter")
plt.xlabel("Quarter")
plt.ylabel("Total Sales")
plt.legend(title='Genre ID')
st.pyplot(fig)
```
